```python
#For bin/exe to png
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
# Define width and height
# Arbitrary at the momment
width, height = 250, 250
```

```python
# Read file using numpy "fromfile()"
# Works with a wide range of file types, converts them into binary format with mode 'rb'
with open('SteamSetup.exe', mode='rb') as f:
    imageData = np.fromfile(f,dtype=np.uint8,count=width*height).reshape(height,width)
```

```python
# Make into PIL Image and save
PILimage = Image.fromarray(imageData)
PILimage.save('result.png')
```

```python
display(Image.open('result.png'))
```
